```python
nigeria = pd.read_csv('covidnig.csv')
```

```python
nigeria['No. Discharged'] = nigeria['No. Discharged'].str.replace(',', '').astype(int)
```

```python
nigeria['No. of Cases (on admission)'] = nigeria['No. of Cases (on admission)'].str.replace(',', '').astype(int)
```

```python
nigeria['No. of Cases (Lab Confirmed)'] = nigeria['No. of Cases (Lab Confirmed)'].str.replace(',', '').astype(int)
```

```python
nigeria.head()
```

```python
no_confirmed = nigeria[['States Affected', 'No. of Cases (Lab Confirmed)']]
```

```python
no_confirmed.head()
```

```python
no_confirmed2 = no_confirmed.nlargest(10, 'No. of Cases (Lab Confirmed)')
```

```python
no_confirmed2
```

```python
no_confirmed2.plot.bar(x='States Affected', y='No. of Cases (Lab Confirmed)')
plt.ylabel('No. of Cases(Lab Confirmed)', size = 20)
plt.title('Top ten states in terms of Confirmed cases', size=24)
plt.subplots(facecolor='w')
```

```python
nig2 = nigeria[['States Affected', 'No. of Cases (on admission)']]
```

```python
nig3 = nig2.nlargest(10, 'No. of Cases (on admission)')
```

```python
nig3
```

```python
nig3.plot.bar(x='States Affected', y='No. of Cases (on admission)')
```
